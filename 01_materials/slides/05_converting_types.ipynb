## Converting between types

Sometimes, we'll get string data that should be treated as a number, or vice versa. We can use the `str()`, `int()` and `float()` functions to convert between data types. If the argument to the type conversion function is not valid in the target data type, Python will produce an error.

## Contents:
1. String to integer
2. String to float
3. Float and int to string

#### String to integer

```python
# int()
# if float --> truncate the decimal part
# if string representing a valid int --> conver the string to an integer
# if string representing a float --> return error
```

```python
int(12.25)
```

```python
int('2')
```

```python
int('2.2222')
```

#### String to float

```python
float(892) # if int --> convert to a floating-point number
```

```python
float('892')
```

```python
float('112.222') # if string representing a valid float --> conver the string to a float
```

```python
float('you')
```

#### Float and int to string
